```python
import math
import numpy as np
```

```python
def fast_powering(base, exponent, modulo):
    pow_limit = math.floor(math.log(exponent, 2))
    a = base
    pow_list = [a]
    for i in range(pow_limit):
        a = a*a
        a = a%modulo
        pow_list.append(a)
    current_exp = exponent - 2**pow_limit
    pow_list.pop(-1)
    pows_left = 2**np.arange(pow_limit-1,-1,-1)
    for x in pows_left:
        if current_exp >= x:
            a = a*pow_list[-1]
            a = a%modulo
            current_exp = current_exp-x
        pow_list.pop(-1)
    return a
```

```python
result = fast_powering(5, 20, 101)
print(result)
```
