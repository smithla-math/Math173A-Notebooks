```python
import numpy as np

def naive_decrypt(Y):
    holder = np.zeros(26)

    for i in range(26):
        X = shift_string(Y, i)
        holder[i] = X.count("E")
        
    # np.argmax finds the location of the maximum value.
    # (If the maximum is obtained multiple times, only the first index is returned.)
    shift_amt = np.argmax(holder)
    X = shift_string(Y, shift_amt)
    
    return (X, shift_amt)
```
