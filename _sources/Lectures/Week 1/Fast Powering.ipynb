# Fast Powering vs Naive Powering

## Naive powering
Take $5^{16}$ by calculating each exponent one at a time

```python
base = 5
exponent = 16
a = base
print(a)
for i in range(exponent):
    a = a * base
    print(a)
```

## Fast Powering

Take $5^{16}$ with successive squaring

```python
base = 5
exponent = 16
a = base
current_exp = 1
print(current_exp, a)
pow_list = [a]
while current_exp < exponent:
    a = a*a
    current_exp *= 2
    pow_list.append(a)
    print(current_exp, a)

print(pow_list)
    
```

## Fast Powering modulo n

Same as above, but we mod by 101

```python
base = 5
exponent = 16
a = base
current_exp = 1
print(current_exp, a)
pow_list = [a]
while current_exp < exponent:
    a = a*a
    a = a%101
    current_exp *= 2
    pow_list.append(a)
    print(current_exp, a)

print(pow_list)
    
```

What if the power we want isn't a power of 2? For instance, what if we want $5^{20}$ instead of $5^{16}$? A clever trick involves writing 20 in base 2, which is equivalent to writing 20 = 16 + 4. This means we can get $5^{20} = 5^{16}\cdot 5^4$, with each power modded down as we go.

```python
pow_list[2]*pow_list[4] #gives 5^16*5^4 = 5^20 mod 101
```

## A general function for fast powering, using successive squaring.

The function below will work for any positive integer power

```python
import math
import numpy as np
```

```python
def fast_powering(base, exponent, modulo):
    pow_limit = math.floor(math.log(exponent, 2))
    a = base
    pow_list = [a]
    for i in range(pow_limit):
        a = a*a
        a = a%modulo
        pow_list.append(a)
    current_exp = exponent - 2**pow_limit
    pow_list.pop(-1)
    pows_left = 2**np.arange(pow_limit-1,-1,-1)
    for x in pows_left:
        if current_exp >= x:
            a = a*pow_list[-1]
            a = a%modulo
            current_exp = current_exp-x
        pow_list.pop(-1)
    return a
```

```python
result = fast_powering(5, 20, 101)
print(result)
```
